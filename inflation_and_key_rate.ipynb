```python
import matplotlib. pyplot as plt
plt.rcParams['figure.figsize'] = [12.0, 6.0]

import okama as ok
```

### Построение графика инфляци за 12 месяцев и ключевой ставки ЦБ

```python
infl = ok.Inflation('RUB.INFL').rolling_inflation
```

```python
cbr = ok.Rate(symbol='RUS_CBR.RATE').values_ts
```

```python
fig = plt.figure(figsize=(12,6))
# fig.subplots_adjust(bottom=0.2, top=1.5)
ax = plt.gca()
ax.plot(cbr.index.to_timestamp(), cbr, color = 'green', linewidth = 2, label='Ключевая ставка ЦБ')
ax.plot(infl.index.to_timestamp(), infl, color = 'blue', linewidth = 2, label='Инфляция за 12 мес')

ax.set_xlim('2009-07', '2021-05')
ax.set_ylim(0.02, 0.175)

leg = ax.legend()
```
