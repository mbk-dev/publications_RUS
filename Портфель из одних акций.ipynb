```python
import matplotlib. pyplot as plt

import okama as ok

%load_ext autoreload
%autoreload 2
```

```python
ls2 = ['SPY.US', 'GC.COMM']
ls3 = ['AGG.US', 'SPY.US', 'GC.COMM']
curr = 'USD'
first_date = '2004-12'
last_date = '2020-10'
```

```python
ef3 = ok.EfficientFrontierReb(symbols=ls3, ccy=curr, first_date=first_date, last_date=last_date, n_points=40, verbose=True, reb_period='year')
```

```python
ef2 = ok.EfficientFrontierReb(symbols=ls2, first_date=first_date, last_date=last_date, ccy=curr, n_points=40, verbose=True, reb_period='year')
```

```python
df2 = ef2.ef_points
```

```python
labels = ['Акции США', 'Золото']

fig = plt.figure(figsize=(12,8))
# fig.subplots_adjust(bottom=0.2, top=1.5)
ok.Plots(symbols=ls2, first_date=first_date, last_date=last_date).plot_assets(kind='cagr', tickers=labels)
ax = plt.gca()
ax.plot(df2.Risk, df2.CAGR)
```
